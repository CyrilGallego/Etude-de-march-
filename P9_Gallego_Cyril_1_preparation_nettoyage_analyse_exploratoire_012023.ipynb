La première importation s'étant bien déroulée, Je créer deux nouveaux dataframe en utilisant la fonction DROP pour ne garder que les champs pertinant à notre analyse.

```python
dispo_alim_1 = dispo_alim.drop (['Code Domaine','Domaine','Code année','Symbole','Description du Symbole'],axis = 1)
dispo_alim_1
```

```python
population_1 = population.drop (['Code Domaine','Domaine','Code année','Symbole','Description du Symbole','Note'],axis = 1)
population_1
```

```python
dispo_alim_1.isnull().sum()
```

Les données de disponibilitée alimentaire n'étant que sur l'années 2017, je créer un filtre pour n'utiliser que les données de population en 2017.

```python
filtre_population_2017 = population_1['Année'] == 2017
population_2017 = population_1[filtre_population_2017]
population_2017
```

Je transforme le champs valeur de mon dataframe population, pour avoir une valeur de population en million d'habitant.
Puis je supprime le champs unté qui n'est plus nécéssaire.

```python
population_2017['Valeur'] = population_2017['Valeur']*1000
population_2017
```

```python
population_2017 = population_2017.drop(['Unité'],axis = 1)
population_2017
```

L'analyse portant sur la vente de volailles, je créer un filtre, sur la disponibilitée alimentaire, ne comprenant que les données en rapport a la viande de volailles.

```python
filtre_volaille = dispo_alim_1['Produit'] == 'Viande de Volailles'
dispo_volaille = dispo_alim_1[filtre_volaille]
dispo_volaille
```

Le champs élément étant remplit de plusieur type de donnée pertianante, j'utilise la fonction PIVOT_TABLE pour créer un dataframe avec pour champs les données du champs élément et je group le tout en focntion du pays.
En second temps on créer un autre dataframe pour ne garder que les champs nécéssaire a l'étude de marché.

```python
dispo_volaille_2 = dispo_volaille.pivot_table(columns = 'Élément',index = ['Zone'],values = 'Valeur',aggfunc = np.sum)
dispo_volaille_2
```

```python
dispo_volaille_3 = dispo_volaille_2[['Exportations - Quantité','Importations - Quantité','Nourriture']]
dispo_volaille_3
```

Supréssion des champs non nécéssaire.

```python
population_2017_2 = population_2017.drop(['Code zone','Code Élément','Élément','Code Produit','Produit','Année'],axis = 1)
population_2017_2
```

Application de la fonction PIVOT_TABLE sur le dataframe de la disponibilité alimentaire sans le filtre volailles et suppression des champs non nécéssaire.

```python
dispo_alim_total = dispo_alim_1.pivot_table(columns = 'Élément',index = ['Zone'],values = 'Valeur',aggfunc = np.sum)
dispo_alim_total
```

```python
dispo_alim_total_2 = dispo_alim_total[['Exportations - Quantité','Importations - Quantité','Nourriture']]
dispo_alim_total_2
```

Je renome les champs pour une meilleur jointure des dataframes.
Jointure de type inner en fonction du champs zone.

```python
dispo_volaille_3.rename(columns={'Exportations - Quantité': 'Exportations Volailles', 
                                   'Importations - Quantité': 'Importation Volailles',
                                   'Nourriture': 'Nourriture Volailles'}, inplace=True)
dispo_volaille_3
```

```python
dispo_alim_total_2.rename(columns={'Exportations - Quantité': 'Exportations Total', 
                                   'Importations - Quantité': 'Importation Total',
                                   'Nourriture': 'Nourriture Total'}, inplace=True)
dispo_alim_total_2
```

```python
étude_marché_volailles = pd.merge(dispo_volaille_3, dispo_alim_total_2, how="inner", on="Zone")
étude_marché_volailles
```

Maintenant création de nouveaux champs (taux import, export et consomation volailles)pour l'analyse.

```python
étude_marché_volailles['Taux export volailles'] = (étude_marché_volailles['Exportations Volailles']/étude_marché_volailles['Exportations Total'])*100
étude_marché_volailles
```

```python
étude_marché_volailles['Taux import volailles'] = (étude_marché_volailles['Importation Volailles']/étude_marché_volailles['Importation Total'])*100
étude_marché_volailles
```

```python
étude_marché_volailles['Taux conso volailles'] = (étude_marché_volailles['Nourriture Volailles']/étude_marché_volailles['Nourriture Total'])*100
étude_marché_volailles
```

Jointure avec le dataframe population.

```python
étude_marché_volailles_2017 = pd.merge(étude_marché_volailles, population_2017_2, how="inner", on="Zone")
étude_marché_volailles_2017
```

Création du champs Valeur en millions d'habitant pour une meilleur lecture du dataframe, puis suppression du champs valeur.

```python
étude_marché_volailles_2017['Valeur en Millions Habitant'] = étude_marché_volailles_2017['Valeur']/1000000
étude_marché_volailles_2017
```

```python
étude_marché_volailles_2017 = étude_marché_volailles_2017.drop(['Valeur'],axis = 1)
étude_marché_volailles_2017
```

Importation des données de stabilité politique, et création du dataframe avec seulement les champs nécéssaire.
Les champs seront renomé avant jointure.

```python
stabilité_politique = pd.read_csv('/Users/user/Desktop/OPENCLASSROOM/MASTER 1 DATA ANALYST/Projet_9/stabilité_politique2017.csv')
stabilité_politique
```

```python
stabilité_politique.info()
```

```python
stabilité_politique_1 = stabilité_politique[['Zone','Valeur']]
stabilité_politique_1
```

```python
stabilité_politique_1.rename(columns={'Valeur': 'Stabilité Politique'
                                          },inplace = True)
stabilité_politique_1
```

Pour une meilleur analyse je créer un champs Taux évolution sur 1O ans, pour cela plusieur sont nécéssaire pour éffectuer le calcule du taux d'évolution.

```python
filtre_population_2007 = population_1['Année'] == 2007
population_2007 = population_1[filtre_population_2007]
population_2007
```

```python
population_2007_2 = population_2007.drop(['Code zone','Code Élément','Élément','Code Produit','Produit','Année','Unité'],axis = 1)
population_2007_2
```

```python
population_2007_2017= pd.merge(population_2007_2, population_2017_2, how="inner", on="Zone")
population_2007_2017
```

```python
population_2007_2017.rename(columns={'Valeur_x': 'Valeur_2007', 
                                   'Valeur_y': 'Valeur_2017'},inplace = True)
population_2007_2017
```

```python
population_2007_2017['Valeur_2017'] = population_2007_2017['Valeur_2017']/1000
population_2007_2017
```

Création du champs Taux évolution population sur 10 ans via les filtres créer précédament, et jointure des données stabilité politiques avec notre dataframe.

```python
étude_marché_volailles_2017['Taux évolution Population 10 ans'] = ((population_2007_2017['Valeur_2017'] - population_2007_2017['Valeur_2007'])/population_2007_2017['Valeur_2007'])*100
étude_marché_volailles_2017
```

```python
étude_marché_volailles_2017_complete= pd.merge(étude_marché_volailles_2017, stabilité_politique_1, how='inner', on='Zone')
étude_marché_volailles_2017_complete
```

Pour terminer le pré-traitement, importation des données sur le PIB 2017, en ne gardant que les champs pertinant.
Je renome les champs pour effectuer une jointure en inner correcte.

```python
pib = pd.read_csv('/Users/user/Desktop/OPENCLASSROOM/MASTER 1 DATA ANALYST/Projet_9/pib2017.csv')
pib
```

```python
pib_1 = pib[['Zone','Valeur']]
pib_1
```

```python
pib_1.rename(columns={'Valeur': 'PIB'
                                          },inplace = True)
pib_1
```

```python
étude_marché_volailles_complete = pd.merge(étude_marché_volailles_2017_complete, pib_1, how='inner', on='Zone')
étude_marché_volailles_complete
```
